### Scraping Top 250 IMDb movies

```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# Define the URL of the site

imdb_url = requests.get('https://www.imdb.com/chart/top/?ref_=nv_mv_250')
```

```python
# get the HTML from the webpage

imdb_url.content
```

```python
# lxml parser

soup = BeautifulSoup(imdb_url.content, 'lxml')
print(soup.prettify())
```

```python
# Find all td tags on the webpage containing the information we want to scrape

imdb_movies = soup.find_all('td', class_ = 'titleColumn')
imdb_movies
```

```python
# movie list

movies_name = []
for mov in imdb_movies:
    mov = mov.get_text().replace('\n', '')
    mov = mov.strip(' ')
    movies_name.append(mov)
movies_name
```

```python
# ratings top 250 imdb

ratings = soup.find_all('td', class_='ratingColumn imdbRating')
ratings
```

```python
# parse ratings top 250 imdb

imdb_ratings = []
for rtng in ratings:
    rtng = rtng.get_text().replace('\n', '')
    rtng = rtng.strip(' ')
    imdb_ratings.append(rtng)
imdb_ratings
```

```python
# create DataFrame

df = pd.DataFrame()
df['Movie Names'] = movies_name
df['IMDB_Ratings'] = imdb_ratings
df.head()
```

```python
# add separate column 'Year'

df['Year'] = df['Movie Names'].apply(lambda title:title.split('(')[-1].replace(')', ''))
df.head()
```

```python
# delate year from 'Movie Names'

list_name = df['Movie Names'].tolist()
new_list = []
for i in list_name:
    i = ''.join(i)
    new_list.append(i[7:-6])
```

```python
df['Movie Names'] = new_list
```

```python
df
```

```python
# swap columns

df = df.reindex(columns=['Movie Names', 'Year', 'IMDB_Ratings'])
```
